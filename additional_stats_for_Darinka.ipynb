Imaš morda kje seznam vseh govorcev v Iriss? In pa morda tudi število besed za vsakega govorca in število besed za vsak posnetek?

# Seznam govorcev in število besed za govorce:

```python
import pandas as pd

pd.set_option("display.max_rows", None)
df = pd.read_csv("snakedir/iriss.csv")
print(df.groupby("who").numw.sum().reset_index().to_html())
```

# Seznam posnetkov in število besed:

```python
print(df.groupby(["speech"]).numw.sum().reset_index().to_html())
```
